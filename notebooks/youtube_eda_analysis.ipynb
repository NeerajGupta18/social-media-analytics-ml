```python
# Strip whitespace from column names
df.columns = df.columns.str.strip()
```

```python
# List of columns to clean and convert
numeric_columns = [
    'subscribers', 'video views', 'uploads', 'video_views_for_the_last_30_days',
    'lowest_monthly_earnings', 'highest_monthly_earnings',
    'lowest_yearly_earnings', 'highest_yearly_earnings',
    'subscribers_for_last_30_days', 'Gross tertiary education enrollment (%)',
    'Population', 'Unemployment rate', 'Urban_population', 'Latitude', 'Longitude'
]

# Clean commas, $ signs, and convert to numeric
for col in numeric_columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype(str).str.replace(',', '').str.replace('$', '').str.strip()
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df['created_date'] = pd.to_datetime(df['created_date'], errors='coerce')
```

```python
# Drop rows where essential data is missing
df = df.dropna(subset=['subscribers', 'video views'])

# Fill remaining missing numeric values with 0
df.loc[:, numeric_columns] = df[numeric_columns].fillna(0)
```

```python
print("Cleaned dataset shape:", df.shape)
print(df.isnull().sum())
df.head(3)
```

```python
# Replace all missing values in the DataFrame with 0
df = df.fillna(0)

# Confirm that all missing values are handled
print("\nMissing values after replacement:\n")
print(df.isnull().sum())
```

### Question 15:

### What is the correlation between the number of subscribers and the population of a country?

```python
country_group = df.groupby('Country').agg({
    'subscribers': 'sum',
    'Population': 'mean'
}).dropna()

correlation = country_group['subscribers'].corr(country_group['Population'])
print("Correlation between Subscribers and Population:", correlation)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.regplot(x='Population', y='subscribers', data=country_group, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.xlabel("Population")
plt.ylabel("Total Subscribers (Per Country)")
plt.title("Correlation between Country Population and YouTube Subscribers")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

INSIGHT : The plot reveals a moderate positive correlation between a countryâs population and the total number of YouTube subscribers. While population size plays a role (more people = more potential viewers), it's not the only factor â tech access, content culture, and platform adoption also heavily influence subscriber growth. Countries like the USA and India skew the trend due to both population and YouTube dominance.

### Question 20:

### What is the average number of subscribers gained per month since the creation of YouTube channels till now?

```python
from datetime import datetime

df['created_date'] = pd.to_datetime(df['created_date'], errors='coerce')
df['months_since_creation'] = ((datetime.now() - df['created_date']).dt.days) // 30
df['months_since_creation'] = df['months_since_creation'].replace(0, 1)
df['avg_subscribers_per_month'] = df['subscribers'] / df['months_since_creation']
print(df['avg_subscribers_per_month'].describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.histplot(df['avg_subscribers_per_month'], bins=40, kde=True, color='dodgerblue')
plt.xlabel("Average Subscribers Gained Per Month")
plt.title("Distribution of Average Monthly Subscriber Growth")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

INSIGHT : The average number of subscribers gained per month varies widely across channels. A few channels show explosive monthly growth, likely due to viral content or brand popularity, while most grow gradually. The distribution is positively skewed, confirming that YouTube subscriber growth follows a power-law pattern â a few dominate, while many grow modestly.
